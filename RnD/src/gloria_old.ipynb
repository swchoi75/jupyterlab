```python
def read_race_file(path):
    df = pd.read_excel(path, sheet_name="Query", skiprows=11)

    df = df.astype({"ConsUnit": str, "Plant": str, "Outlet": str})
    df = df.rename(
        columns={
            "Unnamed: 1": "FS item description",
            "Unnamed: 3": "CU name",
            "Unnamed: 5": "Plant name",
            "Unnamed: 7": "Outlet name",
        }
    )
    df = df.rename(columns=lambda x: re.sub("\nACT", "", x))
    # df = clean_names(df)
    return df
```

```python
race = read_race_file(f_race).clean_names()
race.head(3)
```

```python
supplementary = duckdb.sql("""
SELECT prftctr_pr, prftctr_cc, value_category, value_cat_description, cost_item, cost_item_description,
SUM(eval_per_actual_cos) AS cost, SUM(eval_per_actual_rev) AS sales
FROM ps_db
GROUP BY prftctr_pr, prftctr_cc, value_category, value_cat_description, cost_item, cost_item_description
""")
supplementary.df().head(3)
```

```python
prql_query = """
    from s=supplementary
    join side:left ps_config (s.prftctr_pr == ps_config.profit_center)
"""
options = prql.CompileOptions(format=False)
sql_statement = prql.compile(prql_query, options)
ps_db_config = duckdb.sql(sql_statement)
ps_db_config.df().head(3)
```

```python
prql_query = """
    from p=ps_db_config
    join side:left r=race_gloria(p.cost_item == r.ps_data_ci)
    # filter r.ps_data_ci != null
"""
options = prql.CompileOptions(format=False)
sql_statement = prql.compile(prql_query, options)
project_specific = duckdb.sql(sql_statement)
project_specific.df().head(3)
```

```python
prql_query = """
    from g=gloria
    join side:left r=race_gloria(g.cost_item == r.gl_cost_item)
    filter g.cost_item != "105"
"""
options = prql.CompileOptions(format=False)
sql_statement = prql.compile(prql_query, options)
cc_costs = duckdb.sql(sql_statement)
cc_costs.df().head(3)
```

```python
project_specific.df().to_csv(output_path + "ps.csv")
cc_costs.df().to_csv(output_path + "cc.csv")
```
