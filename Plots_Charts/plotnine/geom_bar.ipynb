```python
import pandas as pd
import numpy as np

from plotnine import (
    ggplot,
    aes,
    geom_bar,
    coord_flip,
    theme_classic
)
from plotnine.data import mpg
```

```python
mpg.head()
```

```python
mpg.to_csv("mpg.csv")
```

```python
ggplot(mpg) + geom_bar(aes(x='class'))
```

```python
ggplot(mpg) + geom_bar(aes(x='class', weight = 'displ'))
```

```python
ggplot(mpg) + geom_bar(aes(x='class', fill='drv'))
```

```python
(
    ggplot(mpg)
    + geom_bar(aes(x='class', fill='drv'))
    + coord_flip()
    + theme_classic()
)
```
