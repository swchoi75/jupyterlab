```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

```python
mpg = sns.load_dataset("mpg")
```

```python
mpg.info()
```

```python
mpg.to_csv("mpg.csv", index=False)
```

```python
sns.relplot(
    data=mpg, kind="line",
    x="horsepower", y="mpg", hue="cylinders",
);
```

```python
sns.lmplot(
    data=mpg,
    x="horsepower", y="mpg", hue="cylinders", 
);
```
